3. a) Lines in the plane
(a) On a single graph, with horizontal axis x and vertical axis y, sketch the two lines y = 2x + 1 and y = 2x + 4. Can you
find a point (x∗, y∗) where they cross one another (read the values as closely as you can from the plot, say to an accuracy
of ±0.5)? If you cannot find a point where the two lines intersect, explain why not?

```julia
import Pkg
Pkg.add("Plots")
using Plots
gr()


```

```julia
y1(x) = 2x + 1
y2(x) = 2x + 4
plot(y1,-5,5)
plot!(y2,-5,5)
```

From the graph we can see that the lines will never meet as they are parallel meaning they have the same slope in this case it is 2, and they have different y intercepts one being 1 and the other being 4.
